```python
import numpy as np
import matplotlib.pyplot as plt

M = float(input("Enter a value for M: "))

x = np.linspace(-10, 10, 100)

y1 = M * x**2
y2 = M * np.sin(x)

plt.figure(figsize=(8, 5))
plt.plot(x, y1, color='blue', linestyle='-', label=f'y = {M}x²')
plt.plot(x, y2, color='red', linestyle='--', label=f'y = {M}sin(x)')

plt.ylim(-5, 50);

plt.legend()
plt.grid(True)
plt.title(f'Plots of y = {M}x² and y = {M}sin(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(12345)

data = np.random.randn(50)

plt.figure(figsize=(10, 6))

plt.subplot(2, 2, 1)
plt.plot(np.cumsum(data), color='green')
plt.title('Cumulative Sum')
plt.xlabel('Index')
plt.ylabel('Cumulative Value')
plt.grid(True)


plt.subplot(2, 2, 2)
x = np.arange(50)
y = data + np.random.normal(scale=0.5, size=50)
plt.scatter(x, y, color='purple')
plt.title('Scatter Plot with Noise')
plt.xlabel('Index')
plt.ylabel('Value')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

dataset_url = 'https://github.com/AnjulaMehto/MCA/raw/main/company_sales_data.csv'
df = pd.read_csv(dataset_url)

# 1. Line plot for total profit of all months
plt.figure(figsize=(8, 5))
sns.lineplot(x=df['month_number'], y=df['total_profit'], marker='o', color='blue')
plt.title('Total Profit Over Months')
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.grid(True)
plt.show()
```

```python
# 2. Multiline plot for all product sales data
plt.figure(figsize=(8, 5))
for column in df.columns[1:-1]:  # Exclude month_number and total_profit
    sns.lineplot(x=df['month_number'], y=df[column], label=column)
plt.title('Product Sales Over Months')
plt.xlabel('Month Number')
plt.ylabel('Sales')
plt.grid(True)
plt.legend()
plt.show()
```

```python

# 3. Bar chart for all features/attributes
plt.figure(figsize=(8, 5))
df.sum().drop('month_number').plot(kind='bar', color='teal')
plt.title('Total Sales and Profit by Feature')
plt.xlabel('Features')
plt.ylabel('Total Values')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
